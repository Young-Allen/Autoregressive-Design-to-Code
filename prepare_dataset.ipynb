```python
from pathlib import Path
from pprint import pprint

import pandas as pd
import pyarrow.parquet as pq
from datasets import load_dataset

DATA_DIR = Path("/mnt/d/datasets/WebSight/data")
assert DATA_DIR.exists(), f"数据目录不存在: {DATA_DIR}"

parquet_files = sorted(DATA_DIR.glob("*.parquet"))
print(f"共找到 {len(parquet_files)} 个 parquet 文件")
for p in parquet_files[:10]:
    print(p.name)
```

```python
def group_files_by_split(files):
    grouped = {}
    for p in files:
        split = p.name.split("-")[0]
        grouped.setdefault(split, []).append(str(p))
    return grouped

files_by_split = group_files_by_split(parquet_files)
{k: len(v) for k, v in files_by_split.items()}
```
